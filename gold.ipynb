```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

## Create the dataset and start clean it to fit the model

Note: 0 value in target it`s mean the price is the same or low and the 1 value in target it`s mean the price will high

```python
#import the dataset and start make it perfect

gold_ds = pd.read_csv("Daily.csv")
gold_ds.drop(['EUR','JPY','GBP','CAD','CHF','INR','CNY','TRY','SAR','IDR','AED','THB','VND','EGP','KRW','RUB','ZAR','AUD'],axis=1,inplace=True)
gold_ds['USD'] = gold_ds['USD'].str.replace(',','').astype(float)
#gold_ds['Date'] = gold_ds['Date'].str.replace('/','-').astype('datetime64[ns]')
#gold_ds['Day'] = gold_ds['Date'].dt.day
#gold_ds['Month'] = gold_ds['Date'].dt.month
#gold_ds['Year'] = gold_ds['Date'].dt.year
gold_ds['Target_bool'] = gold_ds['USD'] > gold_ds['USD'].shift().rolling(1).max()
gold_ds['Target'] = gold_ds['Target_bool'].astype(int)
gold_ds['day1'] = ((gold_ds['USD'].shift(1).fillna(0) - gold_ds['USD'].shift(2).fillna(0)) / gold_ds['USD'].shift(1).fillna(0)) * 100
gold_ds['day2'] = ((gold_ds['USD'].shift(2).fillna(0) - gold_ds['USD'].shift(3).fillna(0)) / gold_ds['USD'].shift(2).fillna(0)) * 100
gold_ds['day3'] = ((gold_ds['USD'].shift(3).fillna(0) - gold_ds['USD'].shift(4).fillna(0)) / gold_ds['USD'].shift(3).fillna(0)) * 100
gold_ds['day4'] = ((gold_ds['USD'].shift(4).fillna(0) - gold_ds['USD'].shift(5).fillna(0)) / gold_ds['USD'].shift(4).fillna(0)) * 100
gold_ds['day5'] = ((gold_ds['USD'].shift(5).fillna(0) - gold_ds['USD'].shift(6).fillna(0)) / gold_ds['USD'].shift(5).fillna(0)) * 100
gold_ds['day6'] = ((gold_ds['USD'].shift(6).fillna(0) - gold_ds['USD'].shift(7).fillna(0)) / gold_ds['USD'].shift(6).fillna(0)) * 100
gold_ds['day7'] = ((gold_ds['USD'].shift(7).fillna(0) - gold_ds['USD'].shift(8).fillna(0)) / gold_ds['USD'].shift(7).fillna(0)) * 100
gold_ds['day8'] = ((gold_ds['USD'].shift(8).fillna(0) - gold_ds['USD'].shift(9).fillna(0)) / gold_ds['USD'].shift(8).fillna(0)) * 100
gold_ds['day9'] = ((gold_ds['USD'].shift(9).fillna(0) - gold_ds['USD'].shift(10).fillna(0)) / gold_ds['USD'].shift(9).fillna(0)) * 100
#gold_ds = gold_ds.fillna(0)
#gold_ds.drop('Date',axis=1)
gold_ds.head(10)
#gold_ds.info()
```

## Create the model and start learning it

```python
from sklearn.model_selection import train_test_split

x = gold_ds.drop(['Date','Target_bool','Target','USD'],axis=1)
y = gold_ds['Target']

x_train , x_test , y_train , y_test = train_test_split(x,y)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
np.random.seed()
for i in range(10, 100, 10):
    print(f"Trying model with {i} estimators...")
    model = RandomForestClassifier(n_estimators=i).fit(x_train, y_train)
    print(f"Model accruacy on test set: {model.score(x_test, y_test)}")
    print("")
```
